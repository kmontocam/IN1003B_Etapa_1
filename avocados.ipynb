```python
import pandas as pd
import numpy as np
import datetime

# Data is measured in thousands of pounds, worldwide
# Therefore, transforming pounds into kg and adjusting to Mexico's export percentage is necessary

poundsToKg = 0.453592
shareMexico = 0.888

# Import volume of avocados imported by the US

df = pd.read_excel(r'TradeTables.xlsx', 'ImpVol')

# Data munging

df = df.iloc[4:10, 3:21].drop(['Unnamed: 5', 'Unnamed: 8', 'Unnamed: 11',
                         'Unnamed: 13', 'Unnamed: 17'], axis = 1)
df = df.set_index('Unnamed: 3')
df.columns = df.iloc[0]
df = df.drop('Market year*')
df = df.sort_index(ascending = True)
df.index.name, df.columns.name = None, None
df = df.stack()
dates = [(month + '-' + year) for year, month in df.index.values]
df.index = pd.to_datetime(dates)
df = df.to_frame(name = "Avocados imported by the US (KG-TONS)")
df.index.name = 'Month'
```

```python
# Avocados imported by the US

df = df * poundsToKg
df
```

```python
# Avocados exported by Mexico to the US (KG-TONS)

df.columns.values[0] = "Avocados exported by Mexico to the US (KG-TONS)"
df = df * shareMexico
df
```
